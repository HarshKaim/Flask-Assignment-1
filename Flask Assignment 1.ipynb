Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Ans. In Flask, the url_for() function is used to build URLs for specific functions or endpoints. This function is particularly useful because it allows you to generate URLs dynamically, based on the function names, rather than hardcoding the URLs directly. This ensures that your URLs are consistent and can be easily modified if the structure of your application changes.

from flask import Flask, url_for, redirect

app = Flask(__name__)

@app.route('/')

def home():

    return 'This is the Home Page'

@app.route('/about')

def about():

    return 'This is the About Page'

@app.route('/user/<username>')

def show_user_profile(username):

    return f'User: {username}'

@app.route('/redirect-to-about')

def redirect_to_about():

    # Redirect to the 'about' page using url_for
    
    return redirect(url_for('about'))

@app.route('/redirect-to-user/<username>')

def redirect_to_user(username):

    # Redirect to the user's profile page using url_for
    
    return redirect(url_for('show_user_profile', username=username))
